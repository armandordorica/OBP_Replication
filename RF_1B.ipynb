```python
def detect_important_variables_for_reward(df, target_var='reward', min_correlation=0.001, top_n=20):
    """
    Detect the most important variables for predicting the target variable (reward).
    
    Analyzes correlations between:
    - action (categorical)
    - position (numeric/categorical)
    - user features (categorical - user_feature_0 through user_feature_9)
    - user-item affinity features (numeric - if present in dataset)
    
    Parameters:
    -----------
    df : pandas.DataFrame
        DataFrame containing all features and the target variable
    target_var : str
        Name of the target variable (default: 'reward')
    min_correlation : float
        Minimum absolute correlation threshold to consider (default: 0.001)
    top_n : int
        Number of top features to display for each category (default: 20)
    
    Returns:
    --------
    dict
        Dictionary containing correlation dataframes for each feature type:
        - 'action': action correlations
        - 'position': position correlations
        - 'user_features': user feature correlations
        - 'affinity_features': user-item affinity correlations (if present)
        - 'summary': overall summary of top correlations across all features
    """
    import pandas as pd
    import numpy as np
    
    results = {}
    all_correlations = []
    
    print("=" * 80)
    print(f"Analyzing Important Variables for Predicting '{target_var}'")
    print("=" * 80)
    print()
    
    # 1. ACTION correlations
    print("1. ACTION Analysis")
    print("-" * 80)
    if 'action' in df.columns:
        action_one_hot = pd.get_dummies(df['action'], prefix='action')
        action_one_hot[target_var] = df[target_var].values
        action_corr = action_one_hot.corr()[target_var].drop(target_var)
        
        action_df = pd.DataFrame({
            'feature_type': 'action',
            'feature_name': 'action',
            'value': action_corr.index.str.replace('action_', ''),
            'correlation': action_corr.values
        })
        action_df = action_df[abs(action_df['correlation']) > min_correlation]
        action_df = action_df.sort_values('correlation', key=abs, ascending=False)
        
        results['action'] = action_df
        all_correlations.append(action_df)
        
        print(f"Found {len(action_df)} action values with |correlation| > {min_correlation}")
        print(f"\nTop {min(top_n, len(action_df))} action correlations:")
        print(action_df.head(top_n).to_string(index=False))
        print()
    else:
        print("'action' column not found in dataframe")
        print()
    
    # 2. POSITION correlations
    print("2. POSITION Analysis")
    print("-" * 80)
    if 'position' in df.columns:
        # Check if position is numeric or categorical
        if pd.api.types.is_numeric_dtype(df['position']):
            # Numeric correlation
            position_corr = df[['position', target_var]].corr().iloc[0, 1]
            position_df = pd.DataFrame({
                'feature_type': ['position'],
                'feature_name': ['position'],
                'value': ['numeric'],
                'correlation': [position_corr]
            })
        else:
            # Categorical - one-hot encode
            position_one_hot = pd.get_dummies(df['position'], prefix='position')
            position_one_hot[target_var] = df[target_var].values
            position_corr = position_one_hot.corr()[target_var].drop(target_var)
            
            position_df = pd.DataFrame({
                'feature_type': 'position',
                'feature_name': 'position',
                'value': position_corr.index.str.replace('position_', ''),
                'correlation': position_corr.values
            })
        
        position_df = position_df[abs(position_df['correlation']) > min_correlation]
        position_df = position_df.sort_values('correlation', key=abs, ascending=False)
        
        results['position'] = position_df
        all_correlations.append(position_df)
        
        print(f"Found {len(position_df)} position values with |correlation| > {min_correlation}")
        print(f"\nTop {min(top_n, len(position_df))} position correlations:")
        print(position_df.head(top_n).to_string(index=False))
        print()
    else:
        print("'position' column not found in dataframe")
        print()
    
    # 3. USER FEATURES correlations
    print("3. USER FEATURES Analysis")
    print("-" * 80)
    user_feature_cols = [col for col in df.columns if col.startswith('user_feature_')]
    
    if user_feature_cols:
        user_features_list = []
        
        for col in user_feature_cols:
            # One-hot encode each user feature
            user_feat_one_hot = pd.get_dummies(df[col], prefix=col)
            user_feat_one_hot[target_var] = df[target_var].values
            user_feat_corr = user_feat_one_hot.corr()[target_var].drop(target_var)
            
            for feature_value, corr in user_feat_corr.items():
                user_features_list.append({
                    'feature_type': 'user_feature',
                    'feature_name': col,
                    'value': feature_value.replace(f'{col}_', ''),
                    'correlation': corr
                })
        
        user_features_df = pd.DataFrame(user_features_list)
        user_features_df = user_features_df[abs(user_features_df['correlation']) > min_correlation]
        user_features_df = user_features_df.sort_values('correlation', key=abs, ascending=False)
        
        results['user_features'] = user_features_df
        all_correlations.append(user_features_df)
        
        print(f"Analyzed {len(user_feature_cols)} user feature columns")
        print(f"Found {len(user_features_df)} user feature values with |correlation| > {min_correlation}")
        print(f"\nTop {min(top_n, len(user_features_df))} user feature correlations:")
        print(user_features_df.head(top_n).to_string(index=False))
        print()
    else:
        print("No user feature columns found in dataframe")
        print()
    
    # 4. USER-ITEM AFFINITY FEATURES correlations
    print("4. USER-ITEM AFFINITY FEATURES Analysis")
    print("-" * 80)
    # Look for columns that might be affinity features (numeric columns that aren't position or reward)
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    exclude_cols = ['position', target_var, 'action']
    affinity_cols = [col for col in numeric_cols if col not in exclude_cols and not col.startswith('user_feature_')]
    
    if affinity_cols:
        affinity_list = []
        
        for col in affinity_cols:
            corr = df[[col, target_var]].corr().iloc[0, 1]
            affinity_list.append({
                'feature_type': 'affinity',
                'feature_name': col,
                'value': 'numeric',
                'correlation': corr
            })
        
        affinity_df = pd.DataFrame(affinity_list)
        affinity_df = affinity_df[abs(affinity_df['correlation']) > min_correlation]
        affinity_df = affinity_df.sort_values('correlation', key=abs, ascending=False)
        
        results['affinity_features'] = affinity_df
        all_correlations.append(affinity_df)
        
        print(f"Analyzed {len(affinity_cols)} affinity feature columns")
        print(f"Found {len(affinity_df)} affinity features with |correlation| > {min_correlation}")
        print(f"\nAffinity feature correlations:")
        print(affinity_df.to_string(index=False))
        print()
    else:
        print("No user-item affinity feature columns found in dataframe")
        print()
    
    # 5. OVERALL SUMMARY
    print("=" * 80)
    print("OVERALL SUMMARY - Top Variables Across All Feature Types")
    print("=" * 80)
    
    if all_correlations:
        summary_df = pd.concat(all_correlations, ignore_index=True)
        summary_df = summary_df.sort_values('correlation', key=abs, ascending=False)
        
        results['summary'] = summary_df
        
        print(f"\nTop {min(top_n, len(summary_df))} most important variables overall:")
        print(summary_df.head(top_n).to_string(index=False))
        print()
        
        # Additional statistics
        print("\nCorrelation Statistics by Feature Type:")
        print("-" * 80)
        for feat_type in summary_df['feature_type'].unique():
            type_df = summary_df[summary_df['feature_type'] == feat_type]
            print(f"\n{feat_type.upper()}:")
            print(f"  Count: {len(type_df)}")
            print(f"  Mean |correlation|: {abs(type_df['correlation']).mean():.4f}")
            print(f"  Max |correlation|: {abs(type_df['correlation']).max():.4f}")
            print(f"  Min |correlation|: {abs(type_df['correlation']).min():.4f}")
    
    return results


# Example usage:
# results = detect_important_variables_for_reward(df_readable, target_var='reward', min_correlation=0.001, top_n=20)
```

```python
# Run the analysis to detect most important variables
results = detect_important_variables_for_reward(
    df_readable, 
    target_var='reward', 
    min_correlation=0.001,  # Filter out very weak correlations
    top_n=20  # Show top 20 features for each category
)

# Access individual results:
# results['action'] - action correlations
# results['position'] - position correlations  
# results['user_features'] - user feature correlations
# results['affinity_features'] - affinity feature correlations (if present)
# results['summary'] - overall summary across all feature types
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print("=" * 80)
print("CROSS-CORRELATION ANALYSIS: One-Hot Encoded User Features")
print("Finding Redundant Feature Values (Full Dataset)")
print("=" * 80)
print()

# Select user features to analyze
user_feature_cols = ['user_feature_0', 'user_feature_1', 'user_feature_2', 'user_feature_3']

# Verify all columns exist
available_cols = [col for col in user_feature_cols if col in df_readable.columns]
print(f"Analyzing features: {available_cols}")
print()

# One-hot encode each user feature
encoded_dfs = []
for col in available_cols:
    encoded = pd.get_dummies(df_readable[col], prefix=col)
    encoded_dfs.append(encoded)
    print(f"{col}: {len(encoded.columns)} unique values")

# Combine all encoded features
all_encoded = pd.concat(encoded_dfs, axis=1)
print(f"\nTotal one-hot encoded features: {len(all_encoded.columns)}")
print()

# Compute correlation matrix on FULL dataset (unconditional)
print("Computing pairwise correlations...")
corr_matrix = all_encoded.corr()

# Find highly correlated pairs (|correlation| > threshold)
correlation_threshold = 0.7
high_corr_pairs = []

feature_names = all_encoded.columns.tolist()
for i, feat_a in enumerate(feature_names):
    for feat_b in feature_names[i+1:]:
        corr_val = corr_matrix.loc[feat_a, feat_b]
        if abs(corr_val) > correlation_threshold:
            high_corr_pairs.append((feat_a, feat_b, corr_val))

# Sort by absolute correlation
high_corr_pairs.sort(key=lambda x: abs(x[2]), reverse=True)

print(f"Found {len(high_corr_pairs)} pairs with |correlation| > {correlation_threshold}")
print()

if len(high_corr_pairs) > 0:
    print("Top 30 Most Correlated Pairs (Potentially Redundant):")
    print("-" * 80)
    for feat_a, feat_b, corr in high_corr_pairs[:30]:
        print(f"{feat_a} <-> {feat_b}: {corr:+.4f}")
    print()
    
    # Visualize correlation heatmap of top correlated features
    top_features = list(set([p[0] for p in high_corr_pairs[:20]] + [p[1] for p in high_corr_pairs[:20]]))
    
    if len(top_features) > 1:
        subset_corr = corr_matrix.loc[top_features, top_features]
        
        mask = np.triu(np.ones_like(subset_corr, dtype=bool))
        
        plt.figure(figsize=(14, 12))
        sns.heatmap(
            subset_corr,
            mask=mask,
            cmap="coolwarm",
            center=0,
            annot=True,
            fmt=".2f",
            linewidths=0.5,
            cbar_kws={"label": "Correlation"},
            vmin=-1,
            vmax=1
        )
        
        plt.title(f"Cross-Correlation: Top Redundant Feature Values (|r| > {correlation_threshold})", 
                  fontsize=14, pad=20)
        plt.xticks(rotation=45, ha='right', fontsize=9)
        plt.yticks(rotation=0, fontsize=9)
        plt.tight_layout()
        plt.show()
else:
    print(f"No feature pairs found with |correlation| > {correlation_threshold}")
    print("Try lowering the threshold to see weaker correlations.")
```

```python
# Compute correlation with reward for ranking redundant features
# This is used for RANKING features when choosing which to keep from redundant pairs
# NOT for detecting redundancy itself

print("=" * 80)
print("CORRELATION WITH REWARD: For Ranking Features")
print("=" * 80)
print()

# Add reward column
all_encoded_with_reward = all_encoded.copy()
all_encoded_with_reward['reward'] = df_readable['reward'].values

# Compute correlation with reward
reward_corr = all_encoded_with_reward.corr()['reward'].drop('reward')
reward_corr_sorted = reward_corr.abs().sort_values(ascending=False)

print("Top 20 features by |correlation| with reward:")
print("-" * 80)
for feat, corr in reward_corr_sorted.head(20).items():
    actual_corr = reward_corr[feat]
    print(f"{feat}: {actual_corr:+.4f}")

print()
print("When removing redundant pairs, keep the feature with HIGHER |correlation| to reward")
```
